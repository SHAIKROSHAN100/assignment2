```python
#string template class

from string import Template
t=Template('value of x is $x')
print(t.substitute({'x':1}))
```

```python
t=Template('Hi I am $name with id number $idn')
print(t.substitute(name="Shaik Roshan", idn="123"))
```

```python
det=[("Prince",21), ('Sai',22) , ('Krishna',23)]
t=Template("$name is $age years old")
for i in det:
    print(t.substitute(name=i[0],age=i[1]))
```

```python

t=Template("$name got $marks in $subject")

print(t.safe_substitute(name="venky",marks=22,subject="English"))

print(t.safe_substitute(name="harry",marks=21)) #safe_substitute

print(t.substitute(name="tom",marks=19)) #substitute
```
