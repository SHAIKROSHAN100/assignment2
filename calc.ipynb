```python
#Create a function of calc by taking two inputs and one input for symbol to choose opearation and perform operation
def calc():
    print("\n*************************CALCULATOR***************************\n")
    
    #Entering the first number 
    num1=int(input("Enter the First Number:"))
    
    #Entering the second number
    num2=int(input("Enter the Second Number"))
    
    #creating the file
    file=open("result.txt","a")
    
    #Defining operations and their symbols
    print("\n SYMBOLS : OPERATIONS \n\n + : SUM\n - : SUBTRACTION\n * : MULTIPLICATION\n / : DIVISION\n")
    
    choice=input("\nEnter the symbol(+, -, *, /):")
    
    #if we print + operation
    if(choice == "+"):
        print("\nThe Sum of two numbers is:", num1+num2)
        file.write(f"\nThe sum of two numbers is: {num1+num2}")
        print("\nData is recorded successfully")
        
    elif(choice=="-"):  
        print("\nThe Subtraction of two numbers is:", num1-num2)
        file.write(f"\nThe Subtraction of two numbers is: {num1-num2}")
        print("\n Data is recorded successfully")
    
    elif(choice=="*"): 
        print("\nThe Product of two numbers is: ",num1*num2)
        file.write(f"\nThe Product of two numbers is: {num1*num2}")
        print("\nData is recorded successfully")
        
    elif(choice=="/"):   
        print("\nThe Quotient is :",num1/num2)
        file.write(f"\nThe Quotient is : {num1/num2}")
        print("\nData is recorded successfully")
    
    else:
        print("Enter Valid Symbol")
    
    
calc()
```
